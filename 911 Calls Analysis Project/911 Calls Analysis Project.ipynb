##### The Most Common Reason for a 911 Call

```python
df['title'].nunique()
```

```python
df['title'].apply(lambda x : x.split(':')[0])
```

```python
Reason = df['title'].apply(lambda x : x.split(':')[0])
```

```python
df['Reason'] = Reason
```

```python
df.head()
```

```python
df['Reason'].value_counts().head(3)
```

```python
sns.countplot(data = df, x = Reason)
plt.title('Reason for the Call', fontsize = 15)
```

##### Relationships by Time Information of a Call - Day of Week

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
time = df['timeStamp'].iloc[0]
time.hour
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time : time.hour)
df['Month'] = df['timeStamp'].apply(lambda time : time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time : time.day_of_week)
```

```python
dmap = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
df.head()
```

```python
sns.countplot(data = df, x = 'Day of Week', hue = 'Reason')

plt.title('Call Count by Day of Week', fontsize = 15)
plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
```

##### Relationships by Time Information of a Call - Month

```python
sns.countplot(data = df, x = 'Month', hue = 'Reason')

plt.title('Call Count by Month', fontsize = 15)
plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))
# There are missing months
```

```python
byMonth = df.groupby(df['Month']).count()
byMonth.head(10)
```

```python
byMonth['twp'].plot()

plt.title('Call Count by Month', fontsize = 15)
```

```python
monthly = byMonth.reset_index()
```

```python
import warnings
warnings.filterwarnings('ignore') 

sns.lmplot(data = monthly, x = 'Month', y = 'twp')
```

##### Relationships by Time Information of a Call - Date

```python
time = df['timeStamp'].iloc[0]
df['Date'] = df['timeStamp'].apply(lambda time : time.date())
```

```python
df.head()
```

```python
byDate = df.groupby(df['Date']).count()
```

```python
byDate.head()
```

```python
plt.figure(figsize = (10, 5))
byDate['twp'].plot()

plt.title('911 Calls Timeline', fontsize = 15)
```

```python
byTraffic = df[df['Reason'] == 'Traffic'].groupby('Date').count()
```

```python
plt.figure(figsize = (10, 5))
byTraffic['twp'].plot()

plt.title('Traffic Calls Timeline', fontsize = 15)
```

```python
byFire = df[df['Reason'] == 'Fire'].groupby('Date').count()

plt.figure(figsize = (10, 5))
byFire['twp'].plot()

plt.title('Fire Calls Timeline', fontsize = 15)
```

```python
byEMS = df[df['Reason'] == 'EMS'].groupby('Date').count()

plt.figure(figsize = (10, 5))
byEMS['twp'].plot()

plt.title('EMS Calls Timeline', fontsize = 15)
```

##### Relationships by Time Information of a Call - Day of Week & Hour

```python
dayhour = df.groupby(by = ['Day of Week', 'Hour']).count()['Reason'].unstack()

dayhour
```

```python
plt.figure(figsize = (10, 10))
sns.heatmap(data = dayhour, cmap = 'coolwarm')

plt.title('Heatmap of 911 Calls by Day of Week & Hour')
```

```python
sns.clustermap(data = dayhour)

plt.title('Clustermap of 911 Calls by Day of Week & Hour', x = 10)
```

##### Relationships by Time Information of a Call - Day of Week & Month

```python
daymonth = df.groupby(by = ['Day of Week', 'Month']).count()['Reason'].unstack()
daymonth
```

```python
plt.figure(figsize = (10, 10))
sns.heatmap(data = daymonth, cmap = 'viridis')

plt.title('Heatmap of 911 Calls by Day of Week & Month')
```

```python
sns.clustermap(data = daymonth, cmap = 'viridis')

plt.title('Clustermap of 911 Calls by Day of Week & Month', x = 10)
```
