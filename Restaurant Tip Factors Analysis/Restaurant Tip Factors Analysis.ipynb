## Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Visualisation

##### Total Bill Distribution

```python
sns.distplot(tips['total_bill'], kde = False, bins = 40)

plt.title('Total Bill Distribution')
plt.show()
```

```python
sns.kdeplot(tips['total_bill'])
plt.show()
```

##### Relationship Between Total Bill and Tip Amount

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')

plt.title('Total Bill vs Tip')
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')

plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')

plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)

plt.show()
```

##### All Features

```python
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
plt.show()
```

##### Bills vs Sex

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips)

plt.title('Total Bill Amount by Sex')
plt.show()
```

```python
sns.countplot(x = 'sex', data = tips)

plt.title('Bill Counts by Sex')
plt.show()
```

##### Relationships with Day of Week

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)

plt.title('Total Bill Amount by Day of Week')
plt.show()
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')

plt.title('Total Bill by Day of Week & by Smoking Status')
plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
plt.show()
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)

plt.title('Total Bill by Day of Week & by Sex')
plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
plt.show()
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter = True, hue = 'sex', dodge = True)  

plt.title('Total Bill by Day of Week & by Sex')
plt.legend(loc = 'upper left', bbox_to_anchor = (1, 1))
plt.show()
```

##### Correlation

```python
tc = tips.corr(numeric_only = True)
```

```python
sns.heatmap(tc, annot = True, cmap = 'coolwarm')
plt.show()
```

##### Tip Amount by Smoking Status and Service Time

```python
g = sns.FacetGrid(tips, row = 'smoker', col = 'time')
g.map(plt.scatter, 'total_bill', 'tip')

plt.show()
```

##### Total Bill vs Tip Amount

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, hue = 'sex', markers = ['o', 'v'],
          scatter_kws = {'s':100}, palette = 'Accent')

plt.title('Total Bill vs Tip Amount')
plt.show()
```

```python
sns.lmplot(x = 'total_bill', y ='tip', data = tips, col = 'day', hue = 'sex',
          aspect = 0.6, palette = 'Accent')

plt.show()
```
